```python
import duckdb
import os

MINIO_ACCESS_KEY = os.getenv("MINIO_ACCESS_KEY")
MINIO_SECRET_KEY = os.getenv("MINIO_SECRET_KEY")

# 1. Iniciar conexão em memória
con = duckdb.connect()

# 2. Carregar extensões para ler S3 (MinIO)
con.execute("INSTALL httpfs; LOAD httpfs;")

# 3. Configurar acesso ao MinIO (localhost pois você está fora do Docker)
con.execute("SET s3_endpoint='localhost:9000';")
con.execute(f"SET s3_access_key_id={MINIO_ACCESS_KEY};")
con.execute(f"SET s3_secret_access_key={MINIO_SECRET_KEY};") # Use sua senha do .env
con.execute("SET s3_use_ssl=false; SET s3_url_style='path';")

# 4. Ler o arquivo Gold e transformar em DataFrame Pandas
df_produtos = con.execute("SELECT * FROM read_parquet('s3://data-lake/gold/resumo_produtos.parquet')").df()

# 5. Ver o resultado
display(df_produtos)
```
